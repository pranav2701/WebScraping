```python
df = pd.read_csv("C:/Users/prana/Desktop/Amazon Scraping - Sheet1.csv")
df.head()
```

URL = "https://www.amazon.{country}/dp/{asin}"


```python
asin_list = df.iloc[:,2]
country_list = df.iloc[:,3]
print(country_list[3])
```

```python
len = df.shape[0]
len
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import json
```

```python
# Install chrome webdriver from https://chromedriver.chromium.org/

json_out = {}

# driver = webdriver.Chrome('C:\\Users\\prana\\Desktop\\chromedriver.exe')
# driver.get('https://hoopshype.com/salaries/players/')
# driver = webdriver.Chrome(service=Service('C:\\Users\\prana\\Desktop\\chromedriver.exe'))

for i in range(len):
  url = "https://www.amazon." + country_list[i] + "/dp/" + asin_list[i]
  try:
    
    driver = webdriver.Chrome(service=Service('C:\\Users\\prana\\Desktop\\chromedriver.exe'))
    # url = "https://www.amazon.fr/dp/000417870X"
    # print(url)
    
    driver.get(url)

    #bypass cookies by accepting it
    cookie = driver.find_element(by=By.ID, value = "sp-cc-accept")
    ActionChains(driver).click(on_element=cookie).perform()

    title = driver.find_elements(by=By.XPATH, value='//span[@id="productTitle"]')[0]
    # img_URL = driver.find_elements(by=By.XPATH, value='//div[contains(@id,"main-image-container")]')
    price = driver.find_elements(by=By.XPATH, value='//span[@id="price"]')[0]
    details = driver.find_elements(by=By.XPATH, value='//div[contains(@class,"a-expander-content")]')[0]


    json_out[i+1] = {
      "Title" : title.text,
      "price" : price.text,
      "details":details.text
    }

    json_out.update(json_out[i+1])

    driver.close()
    
  except:
    print(url + " not available")
    driver.close()

with open('data.json', 'w') as fp:
  json.dump(json_out, fp,indent=4)
```
